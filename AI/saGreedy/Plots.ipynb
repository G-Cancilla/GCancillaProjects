# 2D Visualization (uses `sog_test.py`)

```python
# 2D sampled input
datax = np.loadtxt("data-2.in")
datax
```

```python
# 2D sampled output - [G(datax), G'(datax[:,0]]), G'(datax[:,1])]
datay = np.loadtxt("sog_test121.txt")
datay
```

```python
# SoG function for data above - G(X)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
ax.scatter(datax[:,0],datax[:,1],datay[:,0],s=2)
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("G(X)")
ax.view_init(45,-30) # Note that angles may need to be changed to visualize properly...
```

```python
# Same as above, but using `surface` - G(X)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
X = datax[:,0].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Y = datax[:,1].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Z = datay[:,0].reshape(X.shape)
ax.plot_surface(X, Y, Z)
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("G(X)")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```

```python
# Gradient - dG(X)/dx[0]
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
X = datax[:,0].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Y = datax[:,1].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Z = datay[:,1].reshape(X.shape)
ax.plot_surface(X, Y, Z, color='purple')
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("dG(X)/dx[0]")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```

```python
# Gradient - dG(X)/dx[0]
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
X = datax[:,0].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Y = datax[:,1].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Z = datay[:,2].reshape(X.shape)
ax.plot_surface(X, Y, Z, color='purple')
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("dG(X)/dx[1]")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```

### High-dimensional problems...
Note that visualizing G'(X) for a 2D problem is difficult because there are two inputs and two derivatives, therefore it would require the ability to see in 4 dimensions. We are just looking at two perspectives/projections of G'(X) above which can lead to misinterpretation if not handled carefully. More importantly, this is just for a 2D problem. For a 3D problem, we would have 3 inputs and 3 derivatives and therefore need to see in 9 dimensions... this problem is known as the **curse of dimensionality** because the dimensionality of this space is growing *exponentially* relative to the input size. Even the visualizations above for G(X) are also suffering from this issue as our sampling process (the grid of points needed to evaluate the function) is growing *exponentially* relative to the input size as well.
